```python
#selecting features  
X = df[['Total Cases','Total Active Cases', 'Total Recovered', 'Total Deaths']]  
```

```python
# change name mortality, total_cases, total_active_cases, population, density_population, total_deaths
X = X.rename(columns={'Total Cases': 'total_cases', 'Total Active Cases': 'total_active_cases',  'Total Recovered': 'total_recovered', 'Total Deaths': 'total_deaths'})
```

```python
# Elbow method to find the optimal number of clusters
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss, marker='o')
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# Fit the K-Means model with 3 clusters
kmeans = KMeans(n_clusters=3, init='k-means++', random_state=42)
y_kmeans = kmeans.fit_predict(X)
```

```python
# Add new column to the data
X['Cluster'] = y_kmeans
```

```python
# Visualize the clusters 3d (Total Cases, Total Deaths, Total Recovered)
fig = px.scatter_3d(X, x='total_cases', y='total_deaths', z='total_recovered', color='Cluster', opacity=0.7)
fig.show()
```
